```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('data/diabetes.csv')
df
```

```python
X=df.drop(columns=['Outcome'],axis=1)
y=df['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_feature=X.select_dtypes(exclude='object').columns
len(num_feature)
```

```python
numeric_transfrom=StandardScaler()
preprocessor=ColumnTransformer(
    [
        ('StandardScaler',numeric_transfrom,num_feature)
    ],remainder='passthrough'
)
```

```python
X=preprocessor.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
def evaluate_model(true,pred):
    accuracy=accuracy_score(true,pred)
    f1score=f1_score(true,pred)
    pscore=precision_score(true,pred)
    
    return accuracy,f1score,pscore
```

```python
models={
    'LogisticRegression':LogisticRegression(),
    'DecisionTreeClassifier':DecisionTreeClassifier(),
    'RandomForestClassifier':RandomForestClassifier(),
    'AdaBoostClassifier':AdaBoostClassifier(),
    'GradientBoostingClassifier':GradientBoostingClassifier(),
    'SVC':SVC(),
    'CatBoostClassifier':CatBoostClassifier(),
    'XGBClassifier':XGBClassifier()
}
model_list=[]
accuray_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)
    
    y_train_pred=model.predict(X_train)
    y_test_pred=model.predict(X_test)
    
    y_train_acc,y_train_f1,y_train_pre=evaluate_model(y_train,y_train_pred)
    y_test_acc,y_test_f1,y_test_pre=evaluate_model(y_test,y_test_pred)
    
    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])
    accuray_list.append(y_test_acc)
    
    print("Train performance")
    print('accuracy score:{:.4f}'.format(y_train_acc))
    print('f1 score:{:.4f}'.format(y_train_f1))
    print('precission score :{:.4f}'.format(y_train_pre))
    
    print('\n')
    
    print("Test performance:")
    print('accuracy score:{:.4f}'.format(y_test_acc))
    print('f1 score:{:.4f}'.format(y_test_f1))
    print('precission score:{:.4f}'.format(y_test_pre))
```

```python
pd.DataFrame(list(zip(model_list,accuray_list)),columns=['model_name','accuracy']).sort_values(by=['accuracy'],ascending=False)
```

```python
grad_model=GradientBoostingClassifier()
grad_model.fit(X_train,y_train)
y_pred=grad_model.predict(X_test)
score=accuracy_score(y_test,y_pred)
print("Accuracy of the model is : %.2f", score)
```
